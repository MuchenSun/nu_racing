```python
import cv2
from matplotlib import pyplot as plt
from matplotlib.image import imread, imsave
import numpy as np
```

```python
img = cv2.imread('tech_f2.png')
print(img.shape)
print(img[0][0])
```

```python
count = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j][0] == 255 and img[i][j][1] == 255 and img[i][j][2] == 255:
            count += 1
        else:
            img[i][j][0] = 0
            img[i][j][1] = 0
            img[i][j][2] = 0
print(count)
```

```python
plt.imshow(img)
plt.show()
```
